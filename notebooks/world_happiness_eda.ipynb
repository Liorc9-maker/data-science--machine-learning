```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set_theme()
```

```python
df = pd.read_csv("2015.csv")
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
print("Shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nMissing values:\n", df.isnull().sum())
print("\nSummary statistics:\n", df.describe().T)
```

```python
plt.figure(figsize=(8,5))
df["Happiness Score"].hist(bins=20, edgecolor="black")
plt.xlabel("Happiness Score")
plt.ylabel("Number of Countries")
plt.title("Distribution of Happiness Scores")
plt.show()
```

```python
top10 = df.sort_values("Happiness Score", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x="Happiness Score", y="Country", data=top10, hue="Country", palette="viridis", legend=False)
plt.title("Top 10 Happiest Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Country")
plt.show()
```

```python
bottom10 = df.sort_values("Happiness Score", ascending=True).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x="Happiness Score", y="Country", hue="Country", data=bottom10, palette="magma")
plt.title("Bottom 10 Least Happy Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Country")
plt.show()
```

```python
region_mean = df.groupby("Region")["Happiness Score"].mean().sort_values()

plt.figure(figsize=(10,5))
region_mean.plot(kind="barh", color="skyblue")
plt.xlabel("Average Happiness Score")
plt.title("Average Happiness Score by Region")
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)["Happiness Score"].sort_values(ascending=False)
sns.barplot(x=corr.values, y=corr.index, hue=corr.index, palette="coolwarm")
plt.title("Correlation of Features with Happiness Score")
plt.show()
```

```python
sns.pairplot(df[["Happiness Score", "Economy (GDP per Capita)", "Family", "Health (Life Expectancy)", "Freedom"]])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Economy (GDP per Capita)", y="Happiness Score", hue="Region", data=df, alpha=0.7)
plt.title("Happiness vs Economy (GDP per Capita)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Health (Life Expectancy)", y="Happiness Score", hue="Region", data=df, alpha=0.7)
plt.title("Happiness vs Health (Life Expectancy)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(
    df["Economy (GDP per Capita)"],
    df["Happiness Score"],
    s=df["Health (Life Expectancy)"]*500, # bubble size
    c=df["Freedom"],
    cmap="viridis",
    alpha=0.6,
    edgecolor="k"
)
plt.colorbar(label="Freedom")
plt.xlabel("Economy (GDP per Capita)")
plt.ylabel("Happiness Score")
plt.title("Happiness vs Economy (Bubble = Health, Color = Freedom)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="Generosity", 
    y="Happiness Score", 
    hue="Region",     # color by region
    data=df, 
    alpha=0.7
)
plt.title("Correlation between Generosity and Happiness Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.regplot(
    x="Generosity", 
    y="Happiness Score", 
    data=df, 
    scatter_kws={"alpha":0.6}
)
plt.title("Generosity vs Happiness (with regression line)")
plt.show()
```

```python
corr_value = df["Generosity"].corr(df["Happiness Score"])
print(f"Correlation between Generosity and Happiness Score: {corr_value:.3f}")
```

```python
cols = [
    "Trust (Government Corruption)", 
    "Economy (GDP per Capita)", 
    "Happiness Score", 
    "Dystopia Residual"
]

corr_matrix = df[cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation between Economy, Government Trust, Happiness, and Dystopia Residual")
plt.show()
```

```python
sns.pairplot(df[cols], diag_kind="kde")
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
sns.scatterplot(
    x="Economy (GDP per Capita)", 
    y="Happiness Score", 
    hue="Region", 
    data=df, 
    alpha=0.7
)
plt.title("Economy vs Happiness Score")
plt.show()
```

```python
sns.scatterplot(
    x="Trust (Government Corruption)", 
    y="Happiness Score", 
    hue="Region", 
    data=df, 
    alpha=0.7
)
plt.title("Government Trust vs Happiness Score")
plt.show()
```

```python
sns.scatterplot(
    x="Dystopia Residual", 
    y="Happiness Score", 
    hue="Region", 
    data=df, 
    alpha=0.7
)
plt.title("Dystopia Residual vs Happiness Score")
plt.show()
```
